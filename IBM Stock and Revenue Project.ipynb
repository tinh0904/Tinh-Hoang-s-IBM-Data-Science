```python
import yfinance as yf
```

```python
tesla = yf.Ticker("TSLA")
```

```python
tesla_data = tesla.history(period="max")
tesla_data.head()
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://companiesmarketcap.com/tesla/revenue/"
```

```python
data = requests.get(url).text
```

```python
soup = BeautifulSoup(data, 'html.parser')
```

```python
tesla_revenue = pd.DataFrame(columns=["Year", "Revenue", "Change(%)"])
```

```python
for table in soup.find('tbody').find_all('tr'):
    col = table.find_all('td')
    year = col[0].text
    revenue = col[1].text
    change = col[2].text

    tesla_revenue = pd.concat([tesla_revenue, pd.DataFrame({"Year":[year], "Revenue":[revenue], "Change(%)":[change]})])
```

```python
tesla_revenue.tail()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
def make_graph(tesla_data, columns, title="Tesla Stock Prices Over Time"):
    fig = go.Figure()

    for column in columns:
        fig.add_trace(go.Scatter(x=tesla_data.index, y=tesla_data[column], mode='lines', name=column))

    fig.update_layout(
        title=title,
        xaxis_title='Date',
        yaxis_title='Price',
        legend_title='Price Type',
        height=600,
        width=1000
    )

    fig.show()

columns_to_plot = ['Close', 'Open', 'High', 'Low']
make_graph(tesla_data, columns_to_plot)
```
